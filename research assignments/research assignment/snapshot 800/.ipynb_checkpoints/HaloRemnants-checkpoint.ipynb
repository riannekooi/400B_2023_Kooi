```python
import numpy as np
import astropy.units as u
import astropy.table as tbl
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import array as arr

from ReadFile import Read
from CenterOfMassExtended import CenterOfMass
from ExtendedMassProfile import MassProfile
```

## Finding the Density/Combining

```python
M31_MW_Mass_800 = MassProfile("MW", "M31", 800)
r = np.arange(0.1, 20.0, 0.5)
#using mid points instead of edges for r:
r_without_last = np.delete(r, -1)
r_without_first = np.delete(r, 0)
r_new = ((r_without_last)+(r_without_first))/2
M31_MW_enclosed_density_800 = M31_MW_Mass_800.DensityEnclosed(1, r, volDec=2.0)
```

```python
#plotting data 
plt.semilogy(r_new, M31_MW_enclosed_density_800, label="MW-M31 Enclosed Density", c="maroon", linewidth=3)
plt.xlabel("radius in kpc")
plt.ylabel("log(Density in $M_{sol}$ / volume)")
plt.legend()
plt.title("MW-M31 Enclosed Density Model at Snapshot 800")
plt.savefig("MW-M31 Enclosed Density.pdf")
```

## Hernquist Density Profile


```python
a=40
r = np.arange(0.1, 20.5 , 0.5) # radius range we are using up to 30kpc
M_tot = (1.975*1e12 + 1.921*1e12) # found in HW 3
```

```python
hernquist_density_profile = M31_MW_Mass_800.HernquistDensity(r, a, M_tot)
```

```python
plt.semilogy(r,hernquist_density_profile, '--',label='Hernquist Model', c='midnightblue', linewidth=3)
plt.semilogy(r_new, M31_MW_enclosed_density_800, label="MW-M31 Enclosed Density", c="maroon", linewidth=3)
plt.xlabel('radius in kpc')
plt.ylabel('log(Densityin $M_{sol}$ / volume)')
plt.legend()
plt.title("MW-M31 Hernquist Model")
plt.savefig("Hernquist Model.pdf")
```
